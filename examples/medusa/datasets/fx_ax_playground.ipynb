```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def obj0(x):
    return 0.5*np.sin(1.*x)**2

def obj1(x):
    return 0.5*np.cos(2.*x)

def obj2(x):
    return 0.5*np.cos(2.*x) - 1.*np.exp(-3.*x)

# def obj3(x):
#     return None
```

```python
fig, ax = plt.subplots()

domain = np.linspace(0., 4., 1000)

for ix, obj in enumerate([obj0,obj1,obj2]):
    
    ax.plot(domain, obj(domain), label=f'obj{ix}')
    
ax.legend(loc='lower right')
```

```python
param_opts = [0., 0.5, 1., 1.5, 2., 2.5, 3., 3.5, 4.0]
```

```python
objs = [obj0, obj1, obj2]

def fx(params, Ss, objs):
    fx_val = 0.
    for S_ix, S in enumerate(Ss):
        for Sg in S:
            fx_val += objs[Sg](params[S_ix])
            
    return fx_val
            
            
        
```

```python
param_opts_best = np.array([[0.], [1.5], [0.]])
S_best = [[0], [1], [2]]

fx_val_best = fx(param_opts_best, S_best, objs)
print('fx val best : ', fx_val_best)
```

```python
param_opts_comb = np.array([[0.], [1.5]])
S_comb = [[0], [1, 2]]

fx_val_comb = fx(param_opts_comb, S_comb, objs)
print('fx val comb : ', fx_val_comb)
```
