```python
import olympus
import matplotlib.pyplot as plt
import pickle
import pandas as pd
import seaborn as sns
import numpy as np

```

```python
optimal = max(olympus.Dataset('mmli_opv').data['PredictedSO'])

def plot_trace(data_all_repeats, color, label):
    traces = np.array([d.best_values for d in data_all_repeats]).squeeze()
    mean = np.mean(traces, axis=0)
    stde = np.std(traces, axis=0, ddof=1) / np.sqrt(np.shape(traces)[0] - 1)
    x = np.arange(1, len(mean) + 1, 1)
    plt.axhline(optimal, color='red', linestyle='--')
    plt.plot(x, mean, color=color, label=label)
    plt.fill_between(x, y1=mean-1.96*stde, y2=mean+1.96*stde, alpha=0.2, color=color)
    plt.legend()
    plt.xlabel('# evaluations')
    plt.ylabel('best achieved')

```

```python
####### change these to where you saved the results
atlas_data_all_repeats = pickle.load(open('atlas_ohe/results.pkl', 'rb'))
random_data_all_repeats = pickle.load(open('random/results.pkl', 'rb'))
atlas_desc_data_all_repeats = pickle.load(open('atlas_desc/results.pkl', 'rb'))
atlas_ohe_batched_data_all_repeats = pickle.load(open('atlas_ohe_batched/results.pkl', 'rb'))
atlas_desc_batched_data_all_repeats = pickle.load(open('atlas_desc_batched/results.pkl', 'rb'))
#######
```

```python
plot_trace(atlas_data_all_repeats, 'blue', 'atlas ohe')
plot_trace(random_data_all_repeats, 'red', 'random')
plot_trace(atlas_desc_data_all_repeats, 'green', 'atlas desc')
plt.xlim(0, 200)


plt.savefig('mmli_opv_batch_size_1.png', dpi=300 )
```
