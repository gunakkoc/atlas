```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# res = pickle.load(open('camp_color_mixing_2022-12-5/results.pkl', 'rb'))[1]
res = pickle.load(open('camp_color_mixing_2022-12-5/results.pkl', 'rb'))[1]
print(res)
params = res.observations.get_params()
values = res.observations.get_values()

params = np.concatenate((params, np.array([[0., 1.0, 0.]])))
params
```

```python
fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(
    np.arange(values.shape[0])+1, values, ls='-', marker='D', markersize=10,
)

ax.set_xlabel('# experiments')
ax.set_ylabel(r'L$_2$ distance from target RGB')

plt.tight_layout()
plt.savefig('camp_color_mixing_2022-12-5/obj_results.png', dpi=300)
```

```python
fig, ax = plt.subplots(figsize=(7, 4))

# red, yellow, blue


ax.plot(
    np.arange(params.shape[0])+1, params[:,0], ls='-', marker='D', markersize=8, color='r'
)
ax.plot(
    np.arange(params.shape[0])+1, params[:,1], ls='-', marker='D', markersize=8, color='y'
)
ax.plot(
    np.arange(params.shape[0])+1, params[:,2], ls='-', marker='D', markersize=8, color='b'
)

ax.set_xlabel('# experiments')
ax.set_ylabel('fractional stock solution amount')

plt.tight_layout()
plt.savefig('camp_color_mixing_2022-12-5/params_results.png', dpi=300)
```
